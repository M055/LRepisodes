### LR on YouTube starts at LR 232

```python
import pandas as pd
import numpy as np
import time
```

```python
import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors
```

```python
scopes = ["https://www.googleapis.com/auth/youtube.readonly"]
```

### Authenticate youtube

```python
api_service_name = "youtube"
api_version = "v3"
client_secrets_file = "my_client_secret_file.json"

# Get credentials and create an API client
flow = google_auth_oauthlib.flow.InstalledAppFlow.from_client_secrets_file(client_secrets_file, scopes)
credentials = flow.run_console()
youtube = googleapiclient.discovery.build(
api_service_name, api_version, credentials=credentials)

request = youtube.channels().list(
part="snippet,contentDetails,statistics",
forUsername="GoogleDevelopers"
)
response = request.execute()

print(response)
```

#### Make function to return the video code

```python
def getyoutubeurl(episodename):
    from pandas.io.json import json_normalize
    request = youtube.search().list(
            part="snippet",
            maxResults=2,
            q=episodename
        )
    response = request.execute()
    dum = json_normalize(json_normalize(response)['items'][0])
    youtubeID = dum['id.videoId'][0]
    youtubeurl = 'https://www.youtube.com/watch?v='+youtubeID
    return youtubeurl
```

#### Get YouTube ID for first episode

```python
youtubeurl = getyoutubeurl('Limited Resources - Journey into Nyx Set Review')
```

```python
youtubeurl
```

#### Initialize variables

```python
YouTubeURL = [youtubeurl]
epnum = [231]  # episode number of first youtube video
```

## Get all episode YouTube URLs

```python
#epnummin = 232
epnummin = 232
epnummax = 495
```

```python
for n in np.arange(epnummin,epnummax+1):
    currsearchterm = 'Limited Resources '+str(n)
    youtubeurl = getyoutubeurl(currsearchterm)
    #dum = pd.DataFrame({'YouTubeURL': youtubeurl}, index=[epnum])
    #LRonYouTube.append(dum)   
    YouTubeURL.append(youtubeurl)
    epnum.append(n)
print('Done')
```

```python
LRonYouTube = pd.DataFrame({'YouTubeURL':YouTubeURL}, index=epnum)
LRonYouTube.to_csv('LRonYouTube.csv')
```
