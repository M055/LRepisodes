```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
from pandas.io.json import json_normalize
import re
```

```python
allcontent = pd.read_pickle('lr_eposides_master.pkl')
allcontent.head(3)
```

```python
sunsetlist=[]
epnumlist=[]
for ep in allcontent['episodenum']:
    t = allcontent.loc[allcontent['episodenum']==ep,'title'].tolist()[0]
    c = allcontent.loc[allcontent['episodenum']==ep,'content'].tolist()[0]
    dumt=re.findall('sunset',t.lower())
    dumc=re.findall('sunset',c.lower())
    if (len(dumt)>0) or (len(dumc)>0):
        sunsetlist.append(t)
        epnumlist.append(ep)
sunsetlist_df = pd.DataFrame({'title':sunsetlist}, index=epnumlist)
```

```python
sunsetlist_df.head(3)
```

```python
setrevlist=[]
setrevcontent=[]
epnumlist=[]
for ep in allcontent['episodenum']:
    t = allcontent.loc[allcontent['episodenum']==ep,'title'].tolist()[0]
    c = allcontent.loc[allcontent['episodenum']==ep,'content'].tolist()[0]
    dumts=re.findall('set',t.lower())
    dumtr=re.findall('review',t.lower())
    
    dumcs=re.findall('set',c.lower())
    dumcr=re.findall('review',c.lower())
    
    if (len(dumts)>0 and len(dumtr)>0): #or (len(dumcs)>0 and len(dumcr)>0):
        setrevlist.append(t)
        setrevcontent.append(c)
        epnumlist.append(ep)
setrevlist_df = pd.DataFrame({'title':setrevlist, 'content':setrevcontent}, index=epnumlist)
```

```python
setrevlist_df.head(3)
```

```python
leveluplist=[]
levelupcontent=[]
epnumlist=[]
for ep in allcontent['episodenum']:
    t = allcontent.loc[allcontent['episodenum']==ep,'title'].tolist()[0]
    c = allcontent.loc[allcontent['episodenum']==ep,'content'].tolist()[0]
    dumts=re.findall('level',t.lower())
    dumtr=re.findall('up',t.lower())
    
    dumcs=re.findall('level',c.lower())
    dumcr=re.findall('up',c.lower())
    
    if (len(dumts)>0 and len(dumtr)>0) or (len(dumcs)>0 and len(dumcr)>0):
        leveluplist.append(t)
        levelupcontent.append(c)
        epnumlist.append(ep)
leveluplist_df = pd.DataFrame({'title':leveluplist, 'content':levelupcontent}, index=epnumlist)
```

```python
leveluplist_df.head(3)
```

```python
guestlist=[]
guestcontent=[]
epnumlist=[]
for ep in allcontent['episodenum']:
    t = allcontent.loc[allcontent['episodenum']==ep,'title'].tolist()[0]
    c = allcontent.loc[allcontent['episodenum']==ep,'content'].tolist()[0]
    dumts=re.findall('guest',t.lower())
    dumtr=re.findall('welcome',t.lower())
    
    dumcs=re.findall('guest',c.lower())
    dumcr=re.findall('welcome',c.lower())
    
    if (len(dumts)>0 or len(dumtr)>0) or (len(dumcs)>0 or len(dumcr)>0):
        guestlist.append(t)
        guestcontent.append(c)
        epnumlist.append(ep)
guestlist_df = pd.DataFrame({'title':guestlist, 'content':guestcontent}, index=epnumlist)
```

```python
sunsetlist_df.to_csv('sunsetlist.csv')
setrevlist_df.to_csv('setrevlist.csv')
leveluplist_df.to_csv('leveluplist.csv')
guestlist_df.to_csv('guestlist.csv')
```

## NB: Before running this, manually check the CSV files generated from above and FIX 

```python
sunsetfinal = pd.read_csv('sunsetlistFIXED.csv')
setrevfinal = pd.read_csv('setrevlistFIXED.csv')
levelupfinal = pd.read_csv('leveluplistFIXED.csv')
guestfinal = pd.read_csv('guestlistFIXED.csv')
```

```python
lrmaster = pd.read_csv('lr_eposides_master_05.csv')
lrmaster.head(3)
```

```python
sunsetcol=[]
setrevcol=[]
levelupcol=[]
guestcol=[]
for ep in lrmaster['episodenum']:
    dumss=sunsetfinal.loc[sunsetfinal['epnum']==ep,'sunset'].tolist()
    dumsr=setrevfinal.loc[setrevfinal['epnum']==ep,'setreview'].tolist()
    dumlu=levelupfinal.loc[levelupfinal['epnum']==ep,'levelup'].tolist()
    dumgs=guestfinal.loc[guestfinal['epnum']==ep,'guest'].tolist()
    
    if len(dumss)>0:
        sunsetcol.append(dumss[0])
    else:
        sunsetcol.append(' ')
        
    if len(dumsr)>0:
        setrevcol.append(dumsr[0])
    else:
        setrevcol.append(' ')
    
    if len(dumlu)>0:
        levelupcol.append(dumlu[0])
    else:
        levelupcol.append(' ')
    
    if len(dumgs)>0:
        guestcol.append(dumgs[0])
    else:
        guestcol.append(' ')
```

```python
lrmaster['sunsets']=sunsetcol
lrmaster['setreviews']=setrevcol
lrmaster['levelups']=levelupcol
lrmaster['guests']=guestcol
```
